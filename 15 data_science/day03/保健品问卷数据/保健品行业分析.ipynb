````python
desc = sub_data.describe()
desc
````

````python
corr = sub_data.corr()
corr[corr>0.3] = 9999
corr[corr<=0.3] = 0
corr
````

通过特征相关性分析可得：可以把当前这20道题总结出几个方面作为消费保健品的重要因素：
1. 精神状态
2. 抗衰老
3. 身体健康
4. 外界看法

````python
mean_score = desc.loc['mean'].sort_values(ascending=False)
# 在mean_score 后添加一列，说明每道题目
fields = pd.read_csv('保健品字段介绍.csv', encoding='gbk', header=None, 
                        names=['fid', 'effect', '_'], usecols=['fid', 'effect'],
                        index_col='fid')
fields['mean_score'] = mean_score
fields
````

通过上表可得，多数产品在产品功效上更多关注改善用户的精神状态，其次为抗衰老，再次为身体健康，最后为外界影响。
