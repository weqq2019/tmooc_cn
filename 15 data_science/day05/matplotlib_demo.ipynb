```python
# 基于matplotlib实现可视化
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 基本绘图

```python
x = np.arange(1, 8)
y = np.array([18, 23, 6, 14, 3, 19, 11])
plt.plot(x, y)

# 绘制水平线与垂直线
plt.hlines(10, 2, 6)
plt.vlines(4, 7.5, 17.5)
plt.show()
```

## 绘制正弦函数

```python
# 从-π ~ π 拆分出1000个数
x = np.linspace(-np.pi, np.pi, 1000)
sinx = np.sin(x)  # 求x的正弦函数值
plt.plot(x, sinx, linewidth=2, linestyle='--', color='dodgerblue', alpha=0.9,
        label=r'$y=sin(x)$')

# 绘制cos(x) / 2
cosx = np.cos(x) / 2
plt.plot(x, cosx, linewidth=2, linestyle='-.', color='orangered', alpha=0.9, 
        label=r'$y=\frac{1}{2}cos(x)$')

# 设置可视区域
# plt.xlim(0, np.pi)
# plt.ylim(0, 1)

# 设置x轴坐标刻度
vals = [-np.pi, -np.pi/2, 0, np.pi/2, np.pi]
texts = [r'$-\pi$', r'$-\frac{\pi}{2}$', '0', r'$\frac{\pi}{2}$', r'$\pi$']
plt.xticks(vals, texts)

# 修改坐标轴
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))
plt.yticks([-1, -0.5, 0.5, 1])

# 绘制特殊点
xs = [np.pi/2, np.pi/2]
ys = [0, 1]
plt.scatter(xs, ys, marker='o', s=100, edgecolor='red', facecolor='green', zorder=3)

# 添加备注
plt.annotate(r'$[\frac{\pi}{2}, 1]$', xycoords='data', xy=(np.pi/2, 1),
            textcoords='offset points', xytext=(30, 20),
            fontsize=14, arrowprops=dict(arrowstyle='->', connectionstyle='angle3'))

plt.legend()
plt.show()
```
