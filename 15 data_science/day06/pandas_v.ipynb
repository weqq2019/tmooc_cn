```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
df3 = pd.DataFrame(np.random.randn(1000, 2), 
                   columns=['B', 'C']).cumsum()
# df3.plot(x='B', y='C')
df3['A'] = np.arange(len(df3))
df3.plot(x='A', y='B')
```

```python
ts = pd.Series([1,4,8,3,8], index=['Jan', 'Feb', 'Mar', 'Apr', 'May'])
ts.plot(kind='bar')
```

```python
ts1 = pd.Series([1,4,8,3,8], index=['Jan', 'Feb', 'Mar', 'Apr', 'May'])
ts2 = pd.Series([4,2,6,2,1], index=['Jan', 'Feb', 'Mar', 'Apr', 'May'])
df = pd.DataFrame({'ts1':ts1, 'ts2':ts2})
df.plot.bar()
```

```python
ts.plot.pie(autopct='%.1f%%', figsize=(4, 4))
plt.legend()
```

```python
df.plot.pie(subplots=True, figsize=(8, 4))
```

```python
df.plot.scatter(x='ts1', y='ts2', s=[80, 110, 130, 150, 180], 
                c=[80, 110, 130, 150, 180], cmap='jet')
```
