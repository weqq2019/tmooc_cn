# demo01_ridge.ipynb     岭回归模型

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/Salary_Data2.csv')
x=data['YearsExperience']
y=data['Salary']
plt.scatter(x, y)
```

```python
# 基于普通线性回归，训练模型
import sklearn.linear_model as lm
model = lm.LinearRegression()
x = x.reshape(-1, 1)
model.fit(x, y)
pred_y = model.predict(x)
plt.scatter(x, y)
plt.plot(x, pred_y, color='orangered')


# 基于岭回归构建模型
model = lm.Ridge(80, fit_intercept=True, max_iter=10000)
model.fit(x, y)
pred_y = model.predict(x)
plt.plot(x, pred_y, color='green')
```

## 调整模型超参数的过程

```python
# 设置不同的正则系数，构建不同的模型，使用这些模型对普通的测试样本进行测试。
# 选择评估结果最优秀的模型参数为最终选择。
```

```python
import sklearn.metrics as sm

test_data = np.array([ [3.2, 54445], 
                       [4.5, 61111], 
                       [5.9, 81363], 
                       [8.2, 113812], 
                       [9.6, 112635]])
test_x = test_data[:, 0].reshape(-1, 1)
test_y = test_data[:, 1]
pred_test_y = model.predict(test_x)
print(sm.r2_score(test_y, pred_test_y))
```
